```python
class node:
    def __init__(self,value):
        self.value=value
        self.right=None
        self.left=None
    def insert(root,value) :
        if value<root.value:
            if root.left is None:
                root.left=node(value)
            else:
                insert(root.left,value)     # type: ignore
        else:
             if root.right is None:
                 root.right=node(value)
              
             else:
                 insert(root.right,value)  # type: ignore
 
root = node(55)

values = [25, 75, 15, 60, 90, 35, 10, 30, 65]

for v in values:
    insert(root, v) # type: ignore

print("BST created successfully")
                

               
```
